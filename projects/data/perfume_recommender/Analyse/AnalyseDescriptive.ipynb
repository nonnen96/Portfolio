### **2. Analyse Univariée**
#### **a. Analyse des `Ratings`**

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd


# Charger le fichier Excel
df = pd.read_excel("perfume_data_cleaned (4).xlsx")


# Vérifier et convertir la colonne des ratings en numérique
df['Rating'] = pd.to_numeric(df['Rating'].astype(str).str.replace(',', '.'), errors='coerce')



# Style personnalisé pour un rendu plus esthétique
plt.style.use('dark_background')

# Tracer l'histogramme avec fond noir et bleu nuit pour les barres
sns.histplot(df['Rating'], kde=True, bins=20, color='#1f77b4')
plt.title("Distribution des Ratings des Parfums", color='white')
plt.xlabel("Rating", color='white')
plt.ylabel("Nombre de Parfums", color='white')

# Personnaliser les axes pour correspondre au thème sombre
plt.tick_params(colors='white')
plt.show()
```

Tout d'abord, nous avons voulu travailler sur les avis donnés aux parfums. Nous avons visualisé la distribution des notes à l'aide d'un histogramme. La forme générale de cette distribution est proche d'une courbe gaussienne, mais légèrement décalée vers la gauche, ce qui indique que les évaluations tendent à se concentrer autour des valeurs inférieures à 4. La moyenne de 4 est donc relativement élevée par rapport à l'échelle, ce qui suggère que les utilisateurs ont tendance à attribuer des notes assez positives, mais qu'il y a tout de même une tendance générale à attribuer des évaluations légèrement inférieures à cette moyenne.   
En résumé, bien que la moyenne apparaisse élevée par rapport à l'échelle, elle reflète en fait un biais vers des évaluations positives mais modérées. Cette distribution met en lumière un comportement d'évaluation où la majorité des utilisateurs tendent à attribuer des notes relativement bonnes, mais sans atteindre les extrêmes de l'échelle.

#### **b. Analyse des Genres**

```python
sns.countplot(data=df, x='Gender', hue='Gender', palette='pastel', legend=False)
plt.title("Répartition des Parfums par Genre")
plt.xlabel("Genre")
plt.ylabel("Nombre de Parfums")
plt.show()
```

La répartition du nombre de parfums par genre nous indique que nos données contiennent plus de parfums féminins, information que nous garderons à l'esprit pour la suite.

### **3. Analyse Bivariée**
#### **a. Relation entre `Rating` et `Genre`**

```python
sns.boxplot(data=df, x='Gender', y='Rating', hue='Gender', palette='coolwarm', dodge=False)
plt.title("Ratings par Genre")
plt.xlabel("Genre")
plt.ylabel("Rating")
plt.show()
```

Ce  boxplot montre que les parfums pour hommes ont des notes légèrement supérieures à celles des parfums pour femmes et unisexes. Plusieurs raisons peuvent expliquer cela : les parfums masculins utilisent souvent des accords boisés ou frais, jugés consensuels, tandis que les parfums féminins explorent une palette olfactive plus variée, ce qui peut polariser les avis. De plus, les consommateurs masculins pourraient être moins critiques ou avoir des attentes différentes. Enfin, le marché féminin étant plus saturé, il favorise une plus grande variabilité dans la qualité perçue des produits

#### **b. Popularité (`Rating Count`) et Performance**

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Scatterplot sans les labels en abscisse
sns.scatterplot(data=df, x='Rating Count', y='Rating', hue='Gender', palette='viridis')
plt.title("Popularité vs Rating")
plt.xlabel("Nombre d'Évaluations")
plt.ylabel("Rating")

# Supprimer les ticks en abscisse
plt.xticks([])
# Afficher le graphique
plt.show()
```

Ce nuage de points montre la relation entre le nombre d’évaluations et les notes moyennes des parfums, selon leur genre. La plupart des parfums ont des notes comprises entre 3,5 et 4,5. Les produits très populaires (avec beaucoup d’évaluations) tendent à avoir des notes proches de 4,0, reflétant une certaine homogénéité dans leur appréciation. À l’inverse, les parfums moins populaires présentent des notes plus extrêmes, soit très basses (souvent mal reçus), soit très hautes (appréciés par une niche). Les parfums pour hommes sont globalement stables autour de 4,0, tandis que ceux pour femmes et unisexes montrent une plus grande dispersion. 

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Ajuster la palette en fonction des valeurs réelles
palette = {
    'Men': 'blue',
    'Women': 'purple',
    'Unisex': 'green',
    'Unknown': 'gray'  # Ajoutez cette clé si nécessaire
}

# Convertir 'Rating Count' en numérique (float ou int)
df['Rating Count'] = pd.to_numeric(df['Rating Count'], errors='coerce')

# Créer un graphique sans labels en abscisse
fig, ax = plt.subplots()
sns.scatterplot(data=df, x='Rating Count', y='Rating', hue='Gender', palette=palette, ax=ax)

# Personnaliser le titre et les axes
ax.set_title("Popularité vs Rating")
ax.set_xlabel("Nombre d'Évaluations")
ax.set_ylabel("Rating")
ax.set_xlim(0, df['Rating Count'].max() * 1.1)  # Limite de l'axe x ajustée pour avoir un peu plus d'espace

# Supprimer les ticks en abscisse
ax.tick_params(axis='x', which='both', bottom=False, top=False, labelbottom=False)

# Afficher le graphique
plt.tight_layout()
plt.show()


```

Ce nouveau nuage de point illustre encore mieux le fait que les avis s'uniformisent après un grand nombre d'évaluations. Plus un parfum est célèbre, plus il est évalué et plus il y a de chances pour que les gens le perçoivent bien. A l'inverse, un parfum de niche peut soit être très bien noté, soit très mal reçu. On note également que les parfums de femmes sont les plus soumis à un très grand nombre d'évaluations, renforçant l'idée que les parfums les plus connus sont les parfums féminins. 

### **4. Analyse des Notes Olfactives**
#### **a. Fréquence des Notes Principales**

```python
from collections import Counter
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from matplotlib.colors import LinearSegmentedColormap

# Palette de couleurs personnalisée
colors = ["#4A90E2", "#50E3C2", "#B8E986", "#F5A623", "#D0021B"]
cmap = LinearSegmentedColormap.from_list("custom_cmap", colors)

# Combiner toutes les notes
all_notes = ' '.join(df['Top Notes'] + ' ' + df['Middle Notes'] + ' ' + df['Base Notes'])

# Compter les fréquences des mots
note_counts = Counter(all_notes.split())

# Exclure les mots "non" et "disponible"
excluded_words = {"Non", "disponible","and"}
filtered_note_counts = {word: count for word, count in note_counts.items() if word not in excluded_words}

# Créer le nuage de mots avec des paramètres améliorés
wordcloud = WordCloud(
    width=1200, 
    height=600,
    background_color='black',
    colormap=cmap,  # Appliquer le dégradé de couleurs
    max_words=200,  # Limiter à 200 mots
    contour_color='white',  # Ajouter un contour
    contour_width=2,  # Épaisseur du contour
).generate_from_frequencies(filtered_note_counts)

# Afficher le nuage de mots
plt.figure(figsize=(12, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title("Nuage de mots : Notes Olfactives", fontsize=20, color='white', pad=20)
plt.tight_layout()
plt.show()
```

Les notes les plus présentes dans les parfums analysés incluent des classiques intemporels et polyvalents. La **vanille** apporte une douceur enveloppante, tandis que le **musk** ajoute une touche sensuelle. Les agrumes, représentés par l'**orange** et la **bergamote**, offrent une fraîcheur éclatante. Enfin, les accords floraux de la **rose** et de la **lavande** confèrent élégance et caractère, rendant ces notes incontournables dans la parfumerie.

#### **b. Répartition des Accords (`Main Accord`)**

```python

all_accords = ' '.join(df['Main Accords'].fillna('')).split(', ')
accord_counts = Counter(all_accords)
most_common_accords = pd.DataFrame(accord_counts.most_common(10), columns=['Accord', 'Count'])

sns.barplot(data=most_common_accords, x='Count', y='Accord', hue='Accord', palette='muted', legend=False)
plt.title("Les Accords les plus récurrents")
plt.xlabel("Nombre d'Apparitions")
plt.ylabel("Accords")
plt.show()
```

```python
df = df[df['Gender'] != 'Unknown']

# Fonction pour compter les accords les plus fréquents pour un genre donné
def most_frequent_accords_by_gender(data, gender, top_n=10):
    gender_data = data[data['Gender'] == gender]
    all_accords = ' '.join(gender_data['Main Accords'].fillna('')).split(', ')
    accord_counts = Counter(all_accords)
    return pd.DataFrame(accord_counts.most_common(top_n), columns=['Accord', 'Count'])

# Analyse pour chaque genre
for gender in df['Gender'].unique():
    most_common_accords = most_frequent_accords_by_gender(df, gender)
    
    # Visualisation
    sns.barplot(data=most_common_accords, x='Count', y='Accord', hue='Accord', palette='muted', legend=False)

    plt.title(f"Les accords les plus fréquents pour le genre {gender}")
    plt.xlabel("Nombre d'apparitions")
    plt.ylabel("Accords")
    plt.show()
```

Les résultats montrent des préférences olfactives distinctes selon le genre, avec certaines similitudes pour les parfums unisexes. Pour les hommes, les arômes les plus courants sont **woody**, **aromatic** et **fresh spicy**, suivis de **warm spicy** et **citrus**, reflétant des préférences pour des notes robustes et épicées. Les femmes, quant à elles, privilégient des arômes **sweet**, **powdery**, **woody** et **floral**, avec une affinité pour des parfums plus doux, sucrés et floraux. Pour les parfums unisexes, les arômes **woody**, **sweet** et **powdery** sont également dominants, avec des touches d'épices, de musc et d'ambre, créant un équilibre entre les notes masculines et féminines. Ces résultats montrent une évolution vers des compositions plus équilibrées et polyvalentes.

### **5. Analyse par Marque (`Brand`)**

```python
# Charger les données
df = pd.read_excel('perfume_data_cleaned (4).xlsx')

# Conversion des colonnes en numérique
df['Rating'] = pd.to_numeric(df['Rating'], errors='coerce')
df['Rating Count'] = pd.to_numeric(df['Rating Count'].astype(str).str.replace(',', '.'), errors='coerce')


# Analyse par marque
brand_analysis = (
    df.groupby('Brand')
    .agg(
        Perfume_Count=('Brand', 'size'),
        Average_Rating=('Rating', 'mean'),
        Total_Rating_Count=('Rating Count', 'sum')
    )
    .reset_index()
)

# Trier les résultats
brand_analysis_sorted_by_perfume_count = brand_analysis.nlargest(10, 'Perfume_Count')
brand_analysis_sorted_by_rating = brand_analysis.nlargest(10, 'Average_Rating')

# Visualisation : Top marques par nombre de parfums
plt.figure(figsize=(10, 6))
sns.barplot(data=brand_analysis_sorted_by_perfume_count, x='Perfume_Count', y='Brand', hue='Brand', palette='muted', legend=False)

plt.title("Top 10 des marques par nombre de parfums")
plt.xlabel("Nombre de parfums")
plt.ylabel("Marques")
plt.show()

# Visualisation : Marques les mieux notées
plt.figure(figsize=(10, 6))
sns.barplot(data=brand_analysis_sorted_by_rating, x='Average_Rating', y='Brand', hue='Brand', palette='muted', legend=False)

plt.title("Top 10 des marques les mieux notées")
plt.xlabel("Note moyenne")
plt.ylabel("Marques")
plt.show()
```

Le premier graphique montre les 10 marques avec le plus grand nombre de parfums dans notre base de données. En tête, **The Dua Brand** et **Avon** adoptent des stratégies très différentes : la première propose des reproductions haut de gamme, tandis qu’Avon cible un large public avec des produits abordables.  

Des marques comme **Zara**, **Bath & Body Works**, et **Victoria's Secret** se concentrent sur la diversité et l’accessibilité, avec des gammes variées mais moins axées sur le luxe. À l’inverse, des marques prestigieuses comme **Guerlain** et **Giorgio Armani** privilégient la qualité à la quantité, confirmant leur positionnement haut de gamme malgré un volume inférieur.  

Enfin, des marques régionales comme **O Boticário**, **Natura**, et **Rasasi** combinent qualité et accessibilité, reflétant des stratégies adaptées à leurs marchés respectifs. Ce classement illustre ainsi l’équilibre entre popularité, accessibilité et prestige dans l’industrie des parfums.

En terme de notation, les marques **SpongeBob Squarepants**, **SCENTICKY.CO** et **Mercedez-Benz** arrivent en tête. Ces marques, qui ne sont pas forcément reconnues dans le domaine de la haute parfumerie, arrivent en tête des parfums les mieux notés pour plusieurs raisons. Tout d'abord, elles offrent souvent un excellent rapport qualité-prix, rendant leurs produits accessibles à un large public, ce qui génère des avis positifs, surtout si la qualité est au rendez-vous. Ces marques répondent également à des attentes spécifiques, comme la recherche de parfums simples, agréables et faciles à porter, qui conviennent à un usage quotidien.
