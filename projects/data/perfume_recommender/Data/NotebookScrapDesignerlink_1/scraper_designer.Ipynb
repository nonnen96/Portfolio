   - La page listant toutes les marques est chargée.
   - Le message de consentement pour les cookies est accepté automatiquement.

2. **Scraping des marques** :
   - Les éléments HTML contenant les liens des marques sont identifiés.

- La fermeture du driver est effectuée pour garantir la libération des ressources.
